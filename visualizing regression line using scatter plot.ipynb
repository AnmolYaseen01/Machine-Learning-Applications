```
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
import pandas as pd

digits = load_digits()

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)

# Instantiate and fit the Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)



# Make predictions on the test set
predictions= model.predict(x_test)
mse=mean_squared_error(y_test,predictions)
print(mse)
```

```
model.score(x_test,predictions)
```

```
# Plotting the training data points
plt.scatter(x_train[:, 0], y_train, color='blue', label='Training Data')
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Training Data Points')
plt.legend()
plt.grid(True)
plt.show()
```

NOTE:

Attempting to create a scatter plot using the entire feature matrix x_train as the x-axis values and y_train as the y-axis values might not be suitable for a scatter plot because x_train likely contains multiple features (columns), and you need to choose a specific feature to plot against y_train.


```
# Plotting the regression line on the test data
plt.scatter(x_test[:, 0], y_test, color='blue', label='Test Data')
plt.plot(x_test[:, 0], predictions, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Regression Line on Test Data')
plt.legend()
plt.grid(True)
plt.show()
```

NOTE:
 To ensures that you are visualizing the relationship between one specific feature and the target values in a way that makes sense for a scatter plot. we must chose the first feature for visualization (x_train[:, 0]). This allows you to create a scatter plot with one feature on the x-axis and the target values (y_train) on the y-axis.
